# Strategy Stress Test

Run regime-conditioned block bootstrap stress testing for strategy returns.

```python
from regime_lab import fit_regime_model, load_market_data, load_strategy_returns, stress_test

market = load_market_data('../examples/sample_market_prices.csv')
returns = load_strategy_returns('../examples/sample_strategy_returns.csv')
regimes = fit_regime_model(market, n_regimes=3, lookback=20, min_train=30)
result = stress_test(returns, regimes, n_sims=250, horizon=20, block_size=5)
result.summary
```
