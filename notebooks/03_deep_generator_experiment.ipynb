# Deep Generator Experiment

Prototype the conditional window generator interface. When PyTorch is unavailable, the implementation falls back to same-regime bootstrap sampling with small perturbations.

```python
import numpy as np
from regime_lab import fit_deep_generator

windows = np.random.default_rng(42).normal(0, 0.01, size=(64, 20))
labels = np.array([0, 1, 2, 3] * 16)
generator = fit_deep_generator(windows, labels, epochs=1)
generator.backend, generator.sample(3, regime=1).shape
```
